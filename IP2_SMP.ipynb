### Problem 10 - 12
- Calculates the average temperature for each month and puts it into a dataframe named **monthly_avg**
- Promotes the index for **monthly_avg** to a column called **month**
- Reorders the **monthly_avg** columns so they appear *['month', 'temp_f']*

```python
monthly_avg = pd.DataFrame()
for i in range(1,13):
    monthly_avg = monthly_avg.append({'month': i, 'temp_f': df['temp_f'].loc[df['date_time'].dt.month == i].mean()}, ignore_index=True)
    
monthly_avg['month'] = monthly_avg['month'].astype(int)
monthly_avg
```

### Problem 13
Renames the **monthly_avg** index so that it has an empty name

```python
monthly_avg.index.name = None 
monthly_avg.style.hide_index()
monthly_avg
```

### Problem 14
Changes each entry in the **monthly_avg.month** column so it shows the name of the month and instead of the number

```python
months = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
def convert_month(num):
    return months[num]

monthly_avg['month'] = list(map(lambda x : convert_month(x), range(1,len(months)+1)))
monthly_avg
```

### Problem 15 & 16
- Gets the average temperature for each hour starting from 00:00:00 to 23:00:00
- Renames the hourly_avg index to 'Hour'

```python
hourly_avg = pd.DataFrame()
for i in range(df['date_time'].dt.hour.max()+1):
    hourly_avg = hourly_avg.append({'Hour': i, 'temp_f': df['temp_f'].loc[df['date_time'].dt.hour == i].mean()}, ignore_index=True)

hourly_avg['Hour'] = hourly_avg['Hour'].astype(int)
hourly_avg
```

### Problem 17
Changes each entry in the hourly_avg index so it shows the name of the standard time (not military time)

```python
times = {0: '12:00AM', 1: '1:00AM', 2: '2:00AM', 3: '3:00AM', 4: '4:00AM', 5: '5:00AM', 6: '6:00AM', 7: '7:00AM', 8: '8:00AM', 9: '9:00AM', 10: '10:00AM', 11: '11:00AM', 12: '12:00PM', 13: '1:00PM', 14: '2:00PM', 15: '3:00PM', 16: '4:00PM', 17: '5:00PM', 18: '6:00PM', 19: '7:00PM', 20: '8:00PM', 21: '9:00PM', 22: '10:00PM', 23: '11:00PM'}
def convert_time(num):
    return times[num]

hourly_avg['Hour'] = list(map(lambda x : convert_time(x), range(0,len(times))))
hourly_avg
```
